```python
user_score = 0
BOTH = 100
ONE = 60
PASS = 100
MERIT = 150
DISTINCTION = 200
DISTINCTION_STAR = 300
GAMES = 500
INTERNSHIP = 300


maths_grade = input("What is your maths grade?").upper()
if maths_grade == "B" or maths_grade == "A" or maths_grade == "A*" or maths_grade == "*":
    print("Well done...")
    user_score = user_score + ONE
computing_grade = input("What is your computing grade?").upper()
if computing_grade == "B" or computing_grade == "A" or computing_grade == "A*" or computing_grade == "*":
    print("Well done...")
    user_score = user_score + ONE

elif computing_grade + maths_grade == True:
    user_score = user_score + BOTH


else:
    print("Sorry, but you are unlikely to be able to continue in this career path.")        
    exit()

while True:
    Course = input("What is your prefered course? A-Level or BTEC?")
    if Course == "A-level" or Course == "A level" or Course == "Alevel":
        print("Okay...")
        user_score = user_score + ONE
        break

    elif Course == "BTEC" or Course == "Btec" or Course == "btec":
        print("Okay...")
        user_score = user_score + ONE
        break
    
    elif Course == "Both" or Course == "both" or Course == "don't mind":
        print("Okay...")
        user_score = user_score + BOTH
        break
    else:
        print("Please enter a valid input.")

while True:
    Predicted = input("What grade do you think you will achieve in the BTEC course? (Be honest)")
    if Predicted == "Pass":
        print("Okay...")
        user_score = user_score + PASS
        break

    elif Predicted == "Merit":
        print("Okay...")
        user_score = user_score + MERIT
        break

    elif Predicted == "Distinction":
        print("Okay...")
        user_score = user_score + DISTINCTION
        break

    elif Predicted == "Distinction*" or Predicted == "Distinction *" or Predicted == "*" or Predicted == "distinction*":
        print("Okay...")
        user_score = user_score + DISTINCTION_STAR
        break

    else:
        print("Please enter a valid input.")

while True:
    Experience = input("Have you made any games for any platform before?").upper()
    if Experience == "YES":
        print("Okay...")
        user_score = user_score + GAMES
        break

    elif Experience == "NO":
        print("Okay...")
        user_score = user_score
        break

    else:
        print("Please enter a valid input.")
        
while True:
    internship = input("Will you have any relevant internship or work experience by the time you apply for a job?").upper()
    if internship == "YES":
        print("Okay...")
        user_score = user_score + INTERNSHIP
        break

    elif internship == "NO":
        print("Okay...")
        user_score = user_score
        break
    
    else:
        print("Please enter a valid input.")

if user_score < 300:
    print("You are somewhat unlikely to be able to continue in this career path.")
    exit()

elif 300 <= user_score <= 1000:
    print("You may be able to continue in this career path, but you may find regular work difficult to find.")
    exit()
    
elif user_score > 1000:
    print("You have a chance in this career path, but be aware it is very competitive.")
    exit()



    




    
```
