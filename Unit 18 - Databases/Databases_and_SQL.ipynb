## Cardinality ##

#### If a column in a database has a high cardinality, that means that data held in that column is repeated very little or maybe not repeated at all.

#### An example is a GENDER column. This has a very LOW cardinality because the value held (e.g. M, F) is repeated many times on individual records.

#### Another example is ACCOUNT NUMBER column. This column would have a VERY HIGH cardinality because the value found should be UNIQUE to one record. (Many people won't have the same ACCOUNT NUMBER).

#### Another example is SURNAME. This column's cardinality would be quite high because there's the potential for lots of repetition. Many people have the same surname. (E.g. Smith, Roberts, Singh).
