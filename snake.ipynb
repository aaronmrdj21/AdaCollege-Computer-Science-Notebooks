# Snake game probability 
---

#### Snake is a game where you add numbers to get the highest score, however if a 1 appears you lose all your points, Below shows 2 ways to try and tatically play this game by using average of the score you will get and the amount of 1's that will appear per round

```python
import random
from statistics import mean, median, mode
import numpy
import toyplot

def snake_score():
    score = 0

    while True:
        a = random.randint(1,6)
        b = random.randint(1,6)
        if a == 1 or b == 1:
            return score

        else:
            score += a+b
            

print("You scored:", snake_score())        
data = [snake_score() for i in range(1000)]
print("1000 rounds of SNAKE = ", data)
print("The mean is:", mean(data))
print("The mode is:", mode(data))
print("The median is:",median(data))
toyplot.bars(numpy.histogram([snake_score() for i in range(1000)],bins=range(100)));
```

```python
import random
from statistics import mean, median, mode

def snake_rolls():
    rolls = 0
    while True:
        a = random.randint(1,6)
        b = random.randint(1,6)

        if a == 1 or b == 1:
            return rolls
        else:
            rolls += 1

print("You rolled:", snake_rolls())
        
data = [snake_rolls() for i in range(1000)]
print("1000 rounds of SNAKE = ", data)
print("The mean is:", mean(data))
print("The mode is:", mode(data))
print("The median is:",median(data))
```
